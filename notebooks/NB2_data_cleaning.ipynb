```python
%%R
#grab csv files
a = paste(getwd(), "/*.csv", sep="")
b = Sys.glob(a)
c = lapply(b, fread)
# Separate datafiles
cpi <- c[[1]]
gold <- c[[2]]
silver <- c[[3]]
```

Working with Gold and CPI

```python
%%R

#matching available time frames
cpi_gold <- merge(cpi,gold, by="Date")
Date = cpi_gold$Date
CPI <- cpi_gold$Close.x
Gold <- cpi_gold$Close.y
```

```python
%%R
#conversion from matrix to DT to DF
cpi_gold_matrix <- cbind(Date, CPI, Gold)
cpi_gold_table <- as.data.table(cpi_gold_matrix)
setkey(cpi_gold_table, Date, CPI, Gold)
DF <- as.data.frame(cpi_gold_table)
```

```python
%%R
#conversion from factor to numeric
Gold_factor <- DF$Gold
CPI_factor <- DF$CPI
Gold_numbers <- as.numeric(levels(Gold_factor)[as.integer(Gold_factor)])
CPI_numbers <- as.numeric(levels(CPI_factor)[as.integer(CPI_factor)])
Date <- levels(DF$Date)[DF$Date]
Date <- as.Date(Date, "%m/%d/%Y")
DF$Date <- Date
DF$Gold <- Gold_numbers
DF$CPI <- CPI_numbers
#can now use numberic values to run tests
```

```python
%%R
entries <- length(DF$Date) - 100
DF2 <- DF[1:entries,]
```

```python
%%R

plot(DF2$Date, DF2$CPI, type="n", xlab="Date", ylab="Values")
lines(DF2$Date, DF2$CPI, col="green")
lines(DF2$Date, DF2$Gold, col="red")
```
